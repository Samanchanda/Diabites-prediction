```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python

df=pd.read_csv(r'/kaggle/input/diabetes-dataset/diabetes.csv')
```

```python
x=df.drop(columns='Outcome',axis=1)
y=df['Outcome']
```

```python
x
```

```python
y
```

```python
# data standarlization
s=StandardScaler()
s.fit(x)
standardlized_data=s.transform(x)
```

```python
print(standardlized_data)
```

```python
x=standardlized_data
y=df['Outcome']
```

```python
x
```

```python
y
```

```python
#train test split
#x_train,y_train,x_test,y_test=train_test_split(x,y,random_state=2,test_size=0.2)
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2,test_size=0.2)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
# model training
# gredientBoosting
gb=GradientBoostingClassifier( loss='log_loss', learning_rate=0.1, n_estimators=500, subsample=1.0,  min_samples_split=5, min_samples_leaf=2, max_depth=10)
gb.fit(x_train,y_train)
```

```python
y_pred=gb.predict(x_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
#random forest
rf = RandomForestClassifier(max_depth=5,n_estimators=500,criterion='gini')
rf.fit(x_train,y_train)
y_pred1 = rf.predict(x_test)
y_predict = rf.predict(x_train)
```

```python
# testing accuracy
accuracy_score(y_pred1,y_test)

```

```python
# training accuracy
accuracy_score(y_predict,y_train)
```

```python
#logisticRegression
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(x_train,y_train)
y_pred2=log.predict(x_test)
accuracy_score(y_pred2,y_test)
```

```python
 # Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt=LogisticRegression()
dt.fit(x_train,y_train)
y_pred3=dt.predict(x_test)
accuracy_score(y_pred3,y_test)
```
